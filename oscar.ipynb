```python
df_male = pd.read_csv('./data/oscar_age_male_raw.csv', skipinitialspace=True, index_col='Index')
```

```python
# Creation de données femmes
latest_female_oscars = pd.DataFrame({
    'year': [2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'age': [28, 60, 44, 50, 63, 44, 60],
    'name': ['Emma Stone', 'Frances McDormand', 'Olivia Colman', 'Renée Zellweger', 'Frances McDormand', 'Jessica Chastain', 'Michelle Yeoh'],
    'movie': ['La La Land', 'Three Billboards Outside Ebbing, Missouri', 'The Favourite', 'Judy', 'Nomadland', 'The Eyes of Tammy Faye', 'Everything Everywhere All at Once']
})
df_female = pd.concat([df_female, latest_female_oscars], ignore_index=True)
```

```python
# Creation de données hommes
latest_male_oscars = pd.DataFrame({
    'year': [2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'age': [41, 59, 37, 45, 83, 53, 54],
    'name': ['Casey Affleck', 'Gary Oldman', 'Rami Malek', 'Joaquin Phoenix', 'Anthony Hopkins', 'Will Smith', 'Brendan Fraser'],
    'movie': ['Manchester by the Sea', 'Darkest Hour', 'Bohemian Rhapsody', 'Joker', 'The Father', 'King Richard', 'The Whale']
})
df_male = pd.concat([df_male, latest_male_oscars], ignore_index=True)
```

```python
df_female
```

```python
# suppression des tabulations
df_female.movie = df_female.movie.str.replace('\t', '')
df_male.movie = df_female.movie.str.replace('\t', '')
```

```python
# Rajout de colonne genre
df_male['gender'] = 'm'
df_female['gender'] = 'f'
concat = pd.concat([df_male,df_female])
```

```python
# Creation d'un nouveau csv homme
df_male.to_csv('./data/oscar_age_male_processed.csv', index=False)
```

```python
# Concaténer les dataframes

resultat = pd.concat([df_female, df_male])
finalresult = resultat.sort_values("year")

# Enregistrer le résultat dans un nouveau fichier CSV
finalresult.to_csv('./data/oscar_age.csv', index=False)
```

```python
oscar_counts = finalresult['name'].value_counts()

# Afficher le top 5
top_5 = oscar_counts.head(5)
print(top_5)
```

```python


top_5_df = pd.DataFrame({'Nom': top_5.index, 'Nombre d\'Oscars': top_5.values})

# Afficher le DataFrame
print("Top 5 des personnes avec le plus grand nombre d'Oscars :\n", top_5_df)

# Créer un graphique à barres pour afficher les cinq personnes avec le plus grand nombre d'Oscars
plt.bar(top_5_df['Nom'], top_5_df['Nombre d\'Oscars'], color='skyblue')

# Ajouter des titres et des étiquettes
plt.title('Top 5 des personnes avec le plus grand nombre d\'Oscars')
plt.xlabel('Nom')
plt.ylabel('Nombre d\'Oscars')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
# Définir les dimensions du graphique
plt.rcParams['figure.figsize'] = (8, 6)

# Calculer l'âge moyen par genre
age_mean_by_gender = finalresult.groupby('gender')['age'].mean()

# Créer un graphique à barres pour afficher l'âge moyen par genre
age_mean_by_gender.plot(kind='bar', color=['pink', 'lightblue'])

# Ajouter des titres et des étiquettes
plt.title('Âge moyen par genre')
plt.xlabel('Genre')
plt.ylabel('Âge moyen')
plt.xticks(rotation=0)
plt.show()
```

```python
finalresult['decade'] = finalresult['year'] // 10 * 10


age_mean_by_decade = finalresult.groupby('decade')['age'].mean()

plt.hist(age_mean_by_decade.index, weights=age_mean_by_decade.values, bins=range(min(age_mean_by_decade.index), max(age_mean_by_decade.index) + 10, 10), edgecolor='black', alpha=0.7, color='blue')
X = np.array(age_mean_by_decade.index).reshape(-1, 1)
y = np.array(age_mean_by_decade.values)
regression = LinearRegression().fit(X, y)
plt.plot(X, regression.predict(X), color='black')


plt.title('Évolution de l\'âge moyen par décennie ')
plt.xlabel('Décennie')
plt.ylabel('Âge moyen')
plt.xticks(range(min(age_mean_by_decade.index), max(age_mean_by_decade.index) + 10, 10))
plt.grid(True)
plt.show()
```

```python

lien = finalresult['gender'].replace({'f': 0, 'm': 1}).corr(finalresult['age'])

print("lien entre le genre et l'âge :", lien)
```

```python

oscar_counts_by_gender = finalresult.groupby(['gender', 'name']).size()

mean_oscar_counts_by_gender = oscar_counts_by_gender.groupby('gender').mean()

print("Nombre moyen d'Oscars remportés par les hommes :", mean_oscar_counts_by_gender['m'])
print("Nombre moyen d'Oscars remportés par les femmes :", mean_oscar_counts_by_gender['f'])
```

```python

oscar_counts_by_decade_and_person = finalresult.groupby(['decade', 'name']).size()

multi_oscar_winners_by_decade = oscar_counts_by_decade_and_person[oscar_counts_by_decade_and_person > 1].groupby('decade').size()

print("Nombre de multi-oscarisés par décennie :\n", multi_oscar_winners_by_decade)
```

```python
name = Katharine Hepburn
finalresult['name'] = finalresult['name'].replace('Katharine Hepburn', 'mlt_nom')

mlt_nom_data = finalresult[finalresult['name'] == 'mlt_nom']

cumulative_oscars = mlt_nom_data.groupby('year').size().cumsum()

plt.plot(cumulative_oscars.index, cumulative_oscars.values, marker='o', linestyle='-')

# Ajouter des titres et des étiquettes
plt.title('Progression cumulée des Oscars de ' + name)
plt.xlabel('Année')
plt.ylabel('Nombre total d\'Oscars')
plt.grid(True)

# Afficher le graphique
plt.show('homo')
```
